###  MicroGrad demo

```python
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from micrograd.engine import Value
from micrograd.nn import MLP
```

```python
np.random.seed(1337)
random.seed(1337)
```

```python
# make up a dataset

from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100, noise=0.1) # X 0= data points, y = labels => X.shape = (100, 2), while y.shape = (100, )


y = y*2 - 1 # make y be -1 or 1
# visualize in 2D
plt.figure(figsize=(5,5))
plt.scatter(X[:,0], X[:,1], c=y, s=20, cmap='jet')
```

```python
# initialize a model 
# The input layer size must match the dimensionality of your data. Since the make_moons data has only two features,
# setting the input layer to any number other than 2 would mismatch the data dimensions and cause an error. 
model = MLP(2, [16, 16, 1]) # 2-layer neural network, each hidden layer having 16 neurons. Inputs are bidimensional.
print(model)
print("number of parameters", len(model.parameters()))
```

```python
# loss function
def loss(batch_size=None):
    
    # inline DataLoader :)
    if batch_size is None:
        Xb, yb = X, y # It takes all the dataset if the batch_size is not specified
    else:
        ri = np.random.permutation(X.shape[0])[:batch_size] # np.random.permutation shuffles the indices and then selects the first batch_size indices.
        Xb, yb = X[ri], y[ri] # ... uses these indices to select random data points and the associated labels
    inputs = [list(map(Value, xrow)) for xrow in Xb] # Creats a value object for each Value object <-- wrapping
    
    # FORWARD PASS: forward the model to get scores 
    scores = list(map(model, inputs)) # model is applied to each data point in inputs to calculate the scores (predictions).
	# This results in a list of predicted scores, which we use to compute the loss.

    
    # svm "max-margin" loss
    losses = [(1 + -yi*scorei).relu() for yi, scorei in zip(yb, scores)]
    data_loss = sum(losses) * (1.0 / len(losses))
    # L2 regularization
    alpha = 1e-4
    reg_loss = alpha * sum((p*p for p in model.parameters()))
    total_loss = data_loss + reg_loss
    
    # also get accuracy
    accuracy = [(yi > 0) == (scorei.data > 0) for yi, scorei in zip(yb, scores)]
    return total_loss, sum(accuracy) / len(accuracy)

total_loss, acc = loss()
print(total_loss, acc)
```

```python
# optimization
for k in range(100):
    
    # forward
    total_loss, acc = loss()
    
    # backward
    model.zero_grad()
    total_loss.backward()
    
    # update (sgd)
    learning_rate = 1.0 - 0.9*k/100
    for p in model.parameters(): # gradient descent
        p.data -= learning_rate * p.grad
    
    if k % 1 == 0:
        print(f"step {k} loss {total_loss.data}, accuracy {acc*100}%")
```

```python
# visualize decision boundary

h = 0.25
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Xmesh = np.c_[xx.ravel(), yy.ravel()]
inputs = [list(map(Value, xrow)) for xrow in Xmesh]
scores = list(map(model, inputs))
Z = np.array([s.data > 0 for s in scores])
Z = Z.reshape(xx.shape)

fig = plt.figure()
plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title("Regularization parameter: 1e-4")
```
